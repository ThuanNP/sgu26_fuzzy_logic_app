## Bài 7

### Đề bài

Xét một mạng cục bộ (LAN) gồm các máy trạm liên kết với nhau bằng giao thức Ethernet
với tốc độ tối đa 10 Mbps. Lưu lượng trên mạng được biểu diễn bằng giá trị đỉnh của tổng băng
thông sử dụng.
Hai tập mờ: “Quiet”, “Congested” được dùng để mô tả mức tải dữ liệu cảm nhận được trên
mạng LAN.
Nếu tập nền rời rạc: $X=\{0,1,2,5,7,9,10\}$ biểu diễn mức sử dụng băng thông (Mbps), thì
membership functions của hai tập mờ Quiet ($(\tilde{Q})$) và Congested ($(\tilde{C})$) được cho trong Hình P2.7.
<center>

![Picture-2.7](p27.png "Hình P2.7")

</center>
Đối với các tập mờ này, hãy tìm bằng đồ họa: phép hợp, phép giao, phần bù của mỗi tập,phép
hiệu.

### Bài làm

Ta có tập nền rời rạc:

$$
X=\{0,1,2,5,7,9,10\}
$$

Gọi:

$$
\tilde{Q}=\text{Quiet}, \qquad \tilde{C}=\text{Congested}
$$

Với mỗi $x \in X$, các phép toán tập mờ được tính theo:

$$
\mu_{\tilde{Q}\cup \tilde{C}}(x)=\max\left(\mu_{\tilde{Q}}(x),\mu_{\tilde{C}}(x)\right)
$$

$$
\mu_{\tilde{Q}\cap \tilde{C}}(x)=\min\left(\mu_{\tilde{Q}}(x),\mu_{\tilde{C}}(x)\right)
$$

$$
\mu_{\overline{\tilde{Q}}}(x)=1-\mu_{\tilde{Q}}(x)
$$

$$
\mu_{\overline{\tilde{C}}}(x)=1-\mu_{\tilde{C}}(x)
$$

Hiệu hai tập mờ được tính bằng giao với phần bù:

$$
\tilde{Q}-\tilde{C}=\tilde{Q}\cap \overline{\tilde{C}}
$$

nên:

$$
\mu_{\tilde{Q}-\tilde{C}}(x)
=
\min\left(\mu_{\tilde{Q}}(x),1-\mu_{\tilde{C}}(x)\right)
$$

Tương tự:

$$
\tilde{C}-\tilde{Q}=\tilde{C}\cap \overline{\tilde{Q}}
$$

nên:

$$
\mu_{\tilde{C}-\tilde{Q}}(x)
=
\min\left(\mu_{\tilde{C}}(x),1-\mu_{\tilde{Q}}(x)\right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================
# BÀI 7: QUIET VÀ CONGESTED
# ============================================

# Tập nền
X = np.array([0,1,2,5,7,9,10])

# Membership functions từ Hình P2.7
Q = np.array([1.0,1.0,0.8,0.4,0.1,0.0,0.0])

C = np.array([0.0,0.0,0.1,0.4,0.8,1.0,1.0])

# ============================================
# Các phép toán tập mờ
# ============================================

Q_union_C = np.maximum(Q,C)

Q_inter_C = np.minimum(Q,C)

Q_bar = 1-Q

C_bar = 1-C

Q_minus_C = np.minimum(Q,C_bar)

C_minus_Q = np.minimum(C,Q_bar)

# ============================================
# Đồ thị tập mờ ban đầu
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    Q,
    marker='o',
    linewidth=2,
    label='Quiet'
)

plt.plot(
    X,
    C,
    marker='o',
    linewidth=2,
    label='Congested'
)

plt.title(
    "Hình P2.7"
)

plt.xlabel(
    "Bandwidth (Mbit/s)"
)

plt.ylabel(
    "Membership value"
)

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phép hợp
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    Q_union_C,
    marker='o',
    linewidth=2,
    label='Phép hợp $\\tilde{Q}\\cup \\tilde{C}$'
)

plt.title(
    "Phép hợp $\\tilde{Q}\\cup \\tilde{C}$"
)

plt.xlabel(
    "Bandwidth (Mbit/s)"
)

plt.ylabel(
    "Membership value"
)

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phép giao
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    Q_inter_C,
    marker='o',
    linewidth=2,
    label='Phép giao $\\tilde{Q}\\cap \\tilde{C}$'
)

plt.title(
    "Phép giao $\\tilde{Q}\\cap \\tilde{C}$"
)

plt.xlabel(
    "Bandwidth (Mbit/s)"
)

plt.ylabel(
    "Membership value"
)

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phần bù
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    Q_bar,
    marker='o',
    linewidth=2,
    label='Phần bù $\\overline{\\tilde{Q}}$'
)

plt.plot(
    X,
    C_bar,
    marker='o',
    linewidth=2,
    label='Phần bù $\\overline{\\tilde{C}}$'
)

plt.title(
    "Phần bù $\\overline{\\tilde{Q}}$ và $\\overline{\\tilde{C}}$"
)

plt.xlabel(
    "Bandwidth (Mbit/s)"
)

plt.ylabel(
    "Membership value"
)

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phép hiệu
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    Q_minus_C,
    marker='o',
    linewidth=2,
    label='Phép hiệu $\\tilde{Q}-\\tilde{C}$'
)

plt.plot(
    X,
    C_minus_Q,
    marker='o',
    linewidth=2,
    label='Phép hiệu $\\tilde{C}-\\tilde{Q}$'
)

plt.title(
    "Phép hiệu $\\tilde{Q}-\\tilde{C}$ và $\\tilde{C}-\\tilde{Q}$"
)

plt.xlabel(
    "Bandwidth (Mbit/s)"
)

plt.ylabel(
    "Membership value"
)

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================
# BÀI 8: PHÂN BIỆT ÂM THANH KÍNH VỠ
# ============================================

# Tập nền
X = np.arange(0, 11, 1)

# Tham số
sigma = 2
mu_A = 4     # kính cửa sổ vỡ
mu_B = 8     # ly thủy tinh vỡ

# Membership functions
A = np.exp(-((X - mu_A) ** 2) / (2 * sigma ** 2))
B = np.exp(-((X - mu_B) ** 2) / (2 * sigma ** 2))

# ============================================
# Các phép toán tập mờ
# ============================================

A_union_B = np.maximum(A, B)

A_inter_B = np.minimum(A, B)

A_bar = 1 - A

B_bar = 1 - B

A_minus_B = np.minimum(A, B_bar)

B_minus_A = np.minimum(B, A_bar)


# ============================================
# Bảng kết quả
# ============================================

print("x\tμ_A(x)\tμ_B(x)\tA∪B\tA∩B\tμ_Ā(x)\tμ_B̄(x)\tA-B\tB-A")

for i in range(len(X)):
    print(
        f"{X[i]}\t"
        f"{A[i]:.4f}\t"
        f"{B[i]:.4f}\t"
        f"{A_union_B[i]:.4f}\t"
        f"{A_inter_B[i]:.4f}\t"
        f"{A_bar[i]:.4f}\t"
        f"{B_bar[i]:.4f}\t"
        f"{A_minus_B[i]:.4f}\t"
        f"{B_minus_A[i]:.4f}"
    )

# ============================================
# Đồ thị hai tập mờ ban đầu
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    A,
    marker='o',
    linewidth=2,
    label='$A$- Kính cửa sổ vỡ'
)

plt.plot(
    X,
    B,
    marker='o',
    linewidth=2,
    label='$B$ - Ly thủy tinh vỡ'
)

plt.title("Đồ thị tần số âm thanh kính cửa sổ vỡ và ly thủy tinh vỡ")

plt.xlabel("Tần số x (kHz)")

plt.ylabel("Giá trị thành viên")

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phép hợp
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    A_union_B,
    marker='o',
    linewidth=2,
    label='Phép hợp $A \cup B$'
)

plt.title("Phép hợp $A \cup B$")

plt.xlabel("Tần số x (kHz)")

plt.ylabel("Giá trị thành viên")

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phép giao
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    A_inter_B,
    marker='o',
    linewidth=2,
    label='Phép giao $A \cap B$'
)

plt.title("Phép giao $A \cap B$")

plt.xlabel("Tần số x (kHz)")

plt.ylabel("Giá trị thành viên")

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phần bù
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    A_bar,
    marker='o',
    linewidth=2,
    label='Phần bù $\\bar{A}$'
)

plt.plot(
    X,
    B_bar,
    marker='o',
    linewidth=2,
    label='Phần bù $\\bar{B}$'
)

plt.title("Phần bù $\\bar{A}$ và $\\bar{B}$")

plt.xlabel("Tần số x (kHz)")

plt.ylabel("Giá trị thành viên")

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()

# ============================================
# Đồ thị phép hiệu
# ============================================

plt.figure(figsize=(8,5))

plt.plot(
    X,
    A_minus_B,
    marker='o',
    linewidth=2,
    label='Phép hiệu $A - B$'
)

plt.plot(
    X,
    B_minus_A,
    marker='o',
    linewidth=2,
    label='Phép hiệu $B - A$'
)

plt.title("Phép hiệu $A - B$ và $B - A$")

plt.xlabel("Tần số x (kHz)")

plt.ylabel("Giá trị thành viên")

plt.xticks(X)

plt.ylim(-0.05, 1.05)

plt.grid(True)

plt.legend()

plt.show()
```
